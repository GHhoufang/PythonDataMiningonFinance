```python
#引入必要的工作包
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np

#处理图表内嵌中文字体问题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False 

x = np.arange(100, 201)
y = x*2+1
#设置图片尺寸
plt.figure(figsize=(10,6))
#设置X坐标范围
plt.xlim(100,201)
#绘图
plt.plot(x,y)
#设置坐标轴标注
plt.xlabel('X值',fontsize=16)
plt.ylabel('Y值',fontsize=16)
plt.xticks(fontproperties = 'Times New Roman', size = 14)
plt.yticks(fontproperties = 'Times New Roman', size = 14)
#保存图片到文件
plt.savefig('matplot01.png',dpi=300, bbox_inches='tight')
#显示图片
plt.show()
```

```python
# 导入绘图模块
import matplotlib.pyplot as plt
import numpy as np
#构建数据
Y2016 = [15600,12700,11300,4270,3620]
Y2017 = [17400,14800,12000,5200,4020]
labels = ['Beijing','Shanghai','HongKong','Shenzhen','Guangzhou']
bar_width = 0.5

plt.figure(figsize=(10,6))
#设置条形颜色、透明度、宽度
plt.bar(np.arange(5),Y2016,label='2016',color='yellow',
        alpha=0.8,width=bar_width)
plt.bar(np.arange(5) + bar_width, Y2017, label='2017', 
        color='black',alpha=0.8,width=bar_width)
#添加轴标签
plt.xlabel('Top5 City',fontproperties = 'Times New Roman',fontsize=16)
plt.ylabel('Family Amount',fontproperties = 'Times New Roman',fontsize=16)
plt.xticks(np.arange(5)+bar_width,labels)
plt.ylim([2500,20000])
plt.yticks(fontproperties = 'Times New Roman', size = 14)
plt.xticks(fontproperties = 'Times New Roman', size = 14)
#添加标题
plt.title('Millions Family Amount Top5 City Distribution',
          fontproperties = 'Times New Roman',
         fontsize=16)
#为每个条形图添加数值标签
for x2016,y2016 in enumerate(Y2016):
    plt.text(x2016-0.25,y2016+100,'%s'%y2016)
for x2017,y2017 in enumerate(Y2017):
    plt.text(x2017+0.2, y2017+100, '%s'%y2017)
#显示图例
plt.legend()
plt.savefig('matplot03.png',dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 导入绘图模块
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,6))
#构造数据,调整数据顺序，以免低占比的饼块相邻，无法分辨
edu = [0.2515,0.1057,0.2724,0.2336,0.1368]
labels = ['HSchool','Others','College','Bachelor','Master']
explode = [0,0,0.1,0,0]  # 用于分离突出显示大专学历人群
colors=['#6f6f6f','#9f9f9f','#6f6f6f','#9f9f9f','#cfcfcf'] # 自定义颜色
#将横、纵坐标轴标准化处理，保证饼图是一个正圆，否则为椭圆
plt.axes(aspect='equal')
# 控制x轴和y轴的范围
plt.xlim(0,4)
plt.ylim(0,4)
# 绘制饼图
plt.pie(x = edu, # 绘图数据
    explode=explode, # 突出显示大专人群
    labels=labels, # 添加教育水平标签
    colors=colors, # 设置饼图的自定义填充色
    autopct='%.1f%%', # 设置百分比的格式，这里保留一位小数
    pctdistance=0.5,  # 设置百分比标签与圆心的距离
    labeldistance = 1.0, # 设置教育水平标签与圆心的距离
    startangle = 180, # 设置饼图的初始角度
    radius = 1.5, # 设置饼图的半径
    counterclock = False, # 是否逆时针，这里设置为顺时针方向
    wedgeprops = {'linewidth': 1.5, 'edgecolor':'green'},# 设置饼图内外边界的属性值
    textprops = {'fontsize':15, 'color':'k'}, # 设置文本标签的属性值
    center = (1.8,1.8), # 设置饼图的原点
    frame = 1)# 是否显示饼图的图框，这里设置显示

#删除x轴和y轴的刻度
plt.xticks(())
plt.yticks(())
#添加图标题
plt.title('XX信用违约客户分析')
plt.legend(loc="upper right",fontsize=10)
plt.savefig('matplot04.png',dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Notebook偶尔会出现Seanborn方法生成的图形在cell窗口中不能正常显示的问题，
#因此可以将生成的图像通过savefig()保存后查看。
#在Spyder中以下代码可以正常显示
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 

plt.figure(figsize=(10,6))
sns.set(palette="muted", color_codes=True)
#产生100个随机数，服从正态分布
rs = np.random.RandomState(10)
d = rs.normal(size=100)
sns.distplot(d, kde=False, color="b")
plt.savefig('sea02.png',dpi=300, bbox_inches='tight')
plt.show()
#分别创建4个子图，其中第一个不显示概率曲线，
#第二个不显示直方矩形、显示rug分布条
#第三个只显示分布曲线，第四个默认显示直方矩形与概率曲线
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.distplot(d, kde=False, color="b", ax=axes[0, 0])
sns.distplot(d, hist=False, rug=True, color="r", ax=axes[0, 1])
sns.distplot(d, hist=False, color="g", kde_kws={"shade": True}, ax=axes[1, 0])
sns.distplot(d, color="m", ax=axes[1, 1])

plt.savefig('sea03.png',dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
from scipy import stats

#构造二维矩阵，mean为均值向量，第一维的均值为0，第二维的均值为1
#cov为协方差矩阵，两维之间的协方差为0.5
mean, cov = [0, 1], [(1, .5), (.5, 1)]
#用随机数构造一个二维矩阵，其中第一列均值为0.第二列均值为1
#两列之间的协方差为0.5
data = np.random.multivariate_normal(mean, cov, 1000)
df = pd.DataFrame(data, columns=['x', 'y'])

plt.figure(figsize=(10,6))
#with sns.axes_style('white'):
g=sns.jointplot(x='x', y='y', data=df, kind='hex', color='k')
g.set_axis_labels('X','Y',size=16)
#Seaborn版本0.11之后不再支持annotate函数.
#g.annotate(stats.pearsonr,fontsize=12)
g.ax_joint.text(-3,4,stats.pearsonr(df['x'],df['y']),fontsize=12)
g.savefig('sea04.png',dpi=300,bbox_inches='tight')
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 

#处理图表内嵌中文字体问题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False 
sns.set_style('whitegrid',{'font.sans-serif':['simhei','Arial']})
#构造一个y=ax+b的方程，其中x取值为0-99，y=2x+1
x = np.arange(0, 100)
y = x*2+1
#給y加上一个均值为0，方差为50的扰动，验证Seaborn的回归分析能力
var=np.random.normal(0,50,100)
y=y+var
df1=pd.DataFrame()
df1['CorX']=x
df1['CorY']=y
#做散点回归图
plt.figure(figsize=(10,6))
g1=sns.jointplot(x='CorX', y='CorY', data=df1, kind='reg', color='k')
#Seaborn版本0.11之后不再支持annotate函数.
#g1.annotate(stats.pearsonr,fontsize=12)
g1.ax_joint.text(0,250,stats.pearsonr(df1['CorX'],df1['CorY']),fontsize=12)
g1.set_axis_labels('X值','Y值',size=16)
g1.savefig('sea05.png',dpi=300,bbox_inches='tight')
```

```python
import numpy as np
import matplotlib.pyplot as plt
#line
x=np.linspace(-np.pi,np.pi,256,endpoint=True)
#定义余弦函数正弦函数
c,s=np.cos(x),np.sin(x)
plt.figure(figsize=(10,6))
#画图，以x为横坐标，以c为纵坐标
#linestyle参数表示线型，请参看附录A  
plt.plot(x,c,color='blue',linestyle='-',label='COS',alpha=0.5)
#“r*”表示数据点为红色星号，请参看附录A  
plt.plot(x,s,'r*',label='SIN')
#增加标题
plt.title('COS & SIN')
ax=plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],
[r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi/2$',r'$+\pi$'])
plt.yticks(np.linspace(-1,1,5,endpoint=True))
for label in ax.get_xticklabels()+ax.get_yticklabels():
    label.set_fontsize(16)
label.set_bbox(dict(facecolor='white',edgecolor='None',alpha=0.2))
#图例显示
plt.legend(loc='lower right')
#显示网格
plt.grid()
#显示范围
plt.fill_between(x,np.abs(x)<0.5,c,c>0.5,color='green',alpha=0.25)
t=1
plt.plot([t,t],[0,np.cos(t)],"y",linewidth=3,linestyle="--")
plt.annotate('cos(1)',xy=(t,np.cos(1)),xycoords='data',xytext=(+10,+30),
             textcoords='offset points',
             arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2'))
#先保存后显示，否则会保存空白文件
plt.savefig('sincos.png', dpi=300,bbox_inches='tight')
plt.show()
```

```python
#导入必要的包，包括Pandas，Numpy，Matplotlib，Seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#处理Seaborn图表内嵌中文字体问题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False 
sns.set_style("whitegrid",{'font.sans-serif':['simhei','Arial']})

#从ch05_stock_600000.csv读入股票数据，数据文件跨度为2007/12/4-2018/11/30
stockSPDB = pd.read_csv("ch05_stock_600000.csv",encoding = 'utf-8')
stockSPDB['收盘价'].plot(legend=True ,figsize=(10,4))
print('600000股票收盘价：')
plt.show()

#从ch05_stock_000001.csv读入股票数据，数据文件跨度为2007/12/4-2018/11/30
stockPAB=pd.read_csv("ch05_stock_000001.csv",encoding = 'utf-8')
stockPAB['收盘价'].plot(legend=True,figsize=(10,4))
print('600000股票收盘价：')
plt.show()

#绘制股票价格波动
print('股票收盘价对比：')
plt.figure(figsize=(10,4))
plt.plot(stockSPDB['收盘价'],'r-',label='600000')
plt.plot(stockPAB['收盘价'],'k--',label='000001')
plt.legend()
plt.savefig('收盘价对比.png',dpi=300, bbox_inches='tight')
plt.show()

#调用Pandas对象pct_change()方法计算每日收益
#这个函数用来计算同colnums两个相邻的数字之间的变化率。
print('股票每日收益对比：')
plt.figure(figsize=(10,4))
stockSPDB['DailyRe'] = stockSPDB['收盘价'].pct_change()
stockPAB['DailyRe'] = stockPAB['收盘价'].pct_change()

plt.plot(stockSPDB['DailyRe'],'r-',label='600000')
plt.plot(stockPAB['DailyRe'],'k:',label='000001')
plt.legend()
plt.show()

#加入第三支股票300104，分析三者股价相关性
print('加入第三支股票，分析三者之间每日收益相关性')
stockLS=pd.read_csv("ch05_stock_300104.csv",encoding = 'utf-8')
stockLS['DailyRe'] = stockLS['收盘价'].pct_change()
dataLA=pd.DataFrame()
dataLA['pctSPDB']=stockSPDB['DailyRe']
dataLA['pctPAB']=stockPAB['DailyRe']
dataLA['pctLS']=stockLS['DailyRe']

sns.PairGrid(dataLA).map(plt.scatter)

plt.figure(figsize=(10,8))
plt.title('三只股票每日收益相关矩阵', y=1.05, size=15)
sns.heatmap(dataLA.corr(),linewidths=0.1,vmax=1.0, square=True,
            linecolor='white', 
            annot=True,annot_kws={'size':20,'weight':'bold', 'color':'white'})
plt.savefig('相关矩阵.png', dpi=300, bbox_inches='tight')
```
